```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import csv
import math
import numpy as np
import sys, os
from matplotlib import pyplot
df = pd.read_csv('result.csv')
X =df.loc[:, "X"]
Y =df.loc[:, "Y"]
Z= df.loc[:, "Z"]
fig=plt.figure()
print(len(X))
print(len(Y))
print(len(Z))
ax = plt.subplot(666, projection='3d')
plt.rcParams["figure.figsize"] = [160., 60.]
ax.scatter(X[:1000], Y[:1000], Z[:1000], c='grey')
ax.scatter(X[1000:2000], Y[1000:2000], Z[1000:2000], c='r')

ax.set_zlabel('Z')  # 坐标轴
ax.set_ylabel('Y')
ax.set_xlabel('X')
plt.show()
```
